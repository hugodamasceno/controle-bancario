```python
from datetime import datetime, timedelta
```

```python
class conta_basica:
    def __init__(self, saldo=0):
        self.saldo = int(saldo)
        self.saque_maximo = 500
        self.saques_diarios = 3
        self.hist_extrato = {'Saldo inicial':int(saldo)}
    
    def saque(self, valor):
        valor = int(valor)

        try:
            if datetime.now().date() < self.ultimo_saque:
                self.saques_diarios = 3
        except:
            None

        if valor <= self.saldo and self.saques_diarios > 0 and valor <= self.saque_maximo:
            self.saldo -= valor
            self.saques_diarios -= 1
            self.hist_extrato[f'Item {len(self.hist_extrato.keys())} - Saque'] = valor
            self.ultimo_saque = datetime.now().date()

        elif self.saques_diarios == 0:
            print('Você já realizou os saques disponíveis para hoje.\nTente novamente amanhã.')
        elif valor>self.saque_maximo:
            print(f'O limite de um saque é R$ {self.saque_maximo}')
        else:
            print('Você não tem saldo suficiente para esta operação.\nTente outro valor.')
    
    def extrato(self):

        print(f"Extrato Bancário --- {datetime.now().strftime('%d-%m-%Y %H:%M:%S')}")
        for i, j in self.hist_extrato.items():
            print(i, ": R$ ", j)
        if len(self.hist_extrato.keys()) == 1:
            print('Não há movimentação para mostrar.')
        print(f'Seu saldo atual é de: R$ {self.saldo:.2f}')
    
    def deposito(self, valor=0):
        try:
            valor = int(valor)
            if valor > 0:
                self.saldo += valor
                self.hist_extrato[f'Item {len(self.hist_extrato.keys())} - Depósito (R$)'] = valor 
            else:
                print('Valor não válido.')
        except ValueError:
            print('Valor não válido.')
    
    def user(self):
        while True:
            opcao = input('Qual operação desejada? (s - saque, e - extrato, d - depósito, q - sair): ')

            if opcao == 's':
                valor = input('Quanto você gostaria de sacar? ')
                self.saque(valor)

            elif opcao == 'e':
                self.extrato()

            elif opcao == 'd':
                valor = input('Quanto você gostaria de depositar? ')
                self.deposito(valor)

            elif opcao == 'q':
                break
            else:
                print("Opção inválida")
```
